````python
class Electrodomestico:
    def __init__(self):
        self.enchufado = False
    
    def esta_enchufado(self):
        if not self.enchufado:
            raise Exception('Alerta ⚠️: El electrodoméstico no está enchufado')
    
    def enchufar(self):
        self.enchufado = True
````

### Clase Jugera + Properties  [1.0 puntos]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` según las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepción.
- No permita agregar más de 3 ingredientes a la bandeja a la vez. Si se entregan más de 3 frutas, se levante una excepción.
- Se compruebe que todos los elementos de la lista sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

````python
class Jugera(Electrodomestico):
    def __init__(self):
        super().__init__()
        self._bandeja = []
    
    @property
    def bandeja(self):
        return self._bandeja
    
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        if not isinstance(nueva_bandeja, list):
            raise Exception("Error ❌: La nueva bandeja debe ser una lista.")
        
        if len(nueva_bandeja) > 3:
            raise Exception("Error ❌: No se pueden agregar más de 3 ingredientes a la bandeja.")
        
        for fruta in nueva_bandeja:
            if not isinstance(fruta, dict) or not all(key in fruta for key in ['nombre', 'tipo', 'color', 'pepas']):
                raise Exception("Error ❌: Todos los elementos de la lista deben ser diccionarios con las llaves 'nombre', 'tipo', 'color' y 'pepas'.")
        
        self._bandeja = nueva_bandeja
    
    def agregar_ingrediente(self, nueva_fruta):
        self.bandeja.append(nueva_fruta)
    
    def listar_ingredientes(self):
        if not self.bandeja:
            print("Bandeja vacía")
        else:
            ingredientes = ', '.join(str(item['nombre']) for item in self.bandeja)
            print(f"Ingredientes en la bandeja: {ingredientes}.")
    
    def preparar_jugo(self):
        self.esta_enchufado()
        
        if not self.bandeja:
            raise Exception('Error ❌: La bandeja no tiene ingredientes.')
        
        for ingrediente in self.bandeja:
            if 'pepas' in ingrediente:
                print('Alerta ⚠️: El jugo puede contener restos de pepas.')
                break
        
        nombres_ingredientes = [item['nombre'] for item in self.bandeja]
        mensaje = f"Jugo de {', '.join(nombres_ingredientes)} listo. 🏖️🥤 Que lo disfrutes!!! 🥤🏖️."
        self.bandeja = []
        
        return mensaje
````

4. ¿De qué forma podemos exportar las librerias instaladas en un virtual environment? Escriba el comando que nos permite realizar esto.  (0.25 ptos)

`Escriba su respuesta acá:` 

Para exportar las librerías instaladas en un virtual environment, se puede ejecutar el siguiente comando en la línea de comandos:

```
pip freeze > requirements.txt
```

Este comando generará un archivo llamado `requirements.txt` que contendrá una lista detallada de todas las librerías instaladas en el entorno virtual, incluyendo sus versiones específicas. Este archivo es ampliamente utilizado para compartir y replicar conjuntos de librerías en otros entornos o con otros desarrolladores.


Si se está utilizando `conda` como gestor de entornos virtuales, el comando adecuado sería:

```
conda list --export > conda_requirements.txt
```

Al ejecutar este comando, se creará un archivo llamado `conda_requirements.txt` que alberga la lista completa de librerías instaladas, junto con sus versiones, en el entorno virtual. Este archivo tiene un formato legible por `conda`, lo que permite reconstruir el mismo entorno en otra máquina.

Posteriormente, para instalar las librerías desde el archivo `requirements.txt` en otro entorno, se puede emplear el siguiente comando:

```
pip install -r requirements.txt
```

Si se está utilizando `conda`, el comando apropiado sería:

```
conda create --name newenv --file conda_requirements.txt
```

Estos comandos facilitan la recreación de un entorno virtual con las mismas librerías y versiones presentes en el entorno original.

**YAML**
