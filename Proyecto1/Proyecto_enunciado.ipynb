```python
class Profiler():
    """
    Clase Profiler para realizar an√°lisis exploratorio de datos (EDA) en un DataFrame de pandas.

    Esta clase ofrece m√©todos para resumir estad√≠sticas, detectar valores at√≠picos (outliers),
    generar visualizaciones, limpiar y escalar datos, realizar an√°lisis de clustering y detectar anomal√≠as.

    Atributos:
    -----------
    data : pd.DataFrame
        El DataFrame de pandas que se analizar√°.
    timestamp : str
        Marca de tiempo que indica cu√°ndo se cre√≥ la instancia de Profiler, usada para crear carpetas de salida.
    output_folder : str
        Nombre de la carpeta de salida donde se guardar√°n todos los resultados y archivos generados por Profiler.

    M√©todos:
    --------
    summarize(variables_of_interest_summarize, criteria_for_outliers)
        Resumen de estad√≠sticas descriptivas y detecci√≥n de outliers.
    plot_vars(variables_of_interest_plot, top_n_categories)
        Visualizaci√≥n de distribuciones y recuentos de variables de inter√©s.
    clean_data(variables_of_interest_clean_data, drop_duplicates, drop_na, impute_na, numeric_cols)
        Limpieza y preprocesamiento b√°sico del conjunto de datos.
    scale(scaling_technique)
        Escalado y normalizaci√≥n de los datos.
    make_clusters(clustering_algorithm, n_clusters_range, plot_elbow)
        Ejecuci√≥n de algoritmos de clustering y evaluaci√≥n de los resultados.
    detect_anomalies(anomaly_detection_algorithm)
        Detecci√≥n de anomal√≠as en el conjunto de datos.
    clearGarbage()
        Eliminaci√≥n de carpetas y archivos temporales generados por la instancia de¬†Profiler.
    """
    # 1
    def __init__(self, data):

        self.data = data
        self.timestamp = datetime.now().strftime('%d-%m-%Y')
        self.output_folder = f'EDA_{self.timestamp}'
        os.makedirs(self.output_folder, exist_ok=True)
        
        
    # 2  
    def summarize(self, variables_of_interest=None, criteria_for_outliers=None):
        """
        Resumen de las estad√≠sticas y caracter√≠sticas de las variables en el DataFrame.

        Genera un archivo de resumen que incluye informaci√≥n sobre el tipo de variable, n√∫mero de valores √∫nicos, 
        porcentaje de valores √∫nicos, n√∫mero de valores nulos, porcentaje de valores nulos, n√∫mero de valores cero, 
        porcentaje de valores cero, n√∫mero de valores negativos, porcentaje de valores negativos y estad√≠sticas descriptivas 
        (si la variable es num√©rica).

        Args:
            variables_of_interest (list or None): Lista de variables a incluir en el resumen. Si es None, se utilizan todas las columnas del DataFrame.
            criteria_for_outliers (tuple or None): Criterio para considerar outliers en variables num√©ricas. 
                Debe ser un par de valores (umbral_m√≠nimo, umbral_m√°ximo). Si es None, no se calculan los outliers.

        Returns:
            None
        """
        #print("Aplicando summarize")
        if variables_of_interest is None:
            variables_of_interest = self.data.columns

        summary_file = os.path.join(self.output_folder, 'summary.txt')
        with open(summary_file, 'w') as file:
            for column in variables_of_interest:
                if column in self.data.columns:
                    file.write(f'Variable: {column}\n')
                    file.write(f'Tipo de Variable: {self.data[column].dtype}\n')
                    file.write(f'N√∫mero de Valores √önicos: {self.data[column].nunique()}\n')
                    file.write(f'Porcentaje de Valores √önicos: {100 * self.data[column].nunique() / len(self.data):.2f}%\n')
                    file.write(f'N√∫mero de Valores Nulos: {self.data[column].isnull().sum()}\n')
                    file.write(f'Porcentaje de Valores Nulos: {100 * self.data[column].isnull().sum() / len(self.data):.2f}%\n')

                    if self.data[column].dtype in ['int64', 'float64']:
                        num_zeros = (self.data[column] == 0).sum()
                        num_negatives = (self.data[column] < 0).sum()
                        
                        file.write(f'N√∫mero de Valores Cero: {num_zeros}\n')
                        file.write(f'Porcentaje de Valores Cero: {100 * num_zeros / len(self.data):.2f}%\n')
                        file.write(f'N√∫mero de Valores Negativos: {num_negatives}\n')
                        file.write(f'Porcentaje de Valores Negativos: {100 * num_negatives / len(self.data):.2f}%\n')
                        
                        # Considerar outliers si se proporciona un criterio
                        if criteria_for_outliers is not None:
                            data_column = self.data[column]
                            min_threshold, max_threshold = criteria_for_outliers
                            num_outliers = ((data_column < min_threshold) | (data_column > max_threshold)).sum()
                            file.write(f'N√∫mero de Outliers: {num_outliers}\n')
                            file.write(f'Porcentaje de Outliers: {100 * num_outliers / len(self.data):.2f}%\n')
                    
                        statistics = self.data[column].describe()
                        file.write('Estad√≠sticas Descriptivas:\n')
                        file.write(statistics.to_string() + '\n')
                    
                    file.write('\n')
                    
    # 3       
    def plot_vars(self, data, variables_of_interest=None, top_n_categories=5):
        """
        Genera visualizaciones para variables en el DataFrame.

        Este m√©todo crea visualizaciones que incluyen gr√°ficos de correlaci√≥n para variables num√©ricas, matrices de correlaci√≥n
        de V de Cramer para variables categ√≥ricas y gr√°ficos espec√≠ficos para cada variable, ya sea num√©rica o categ√≥rica.

        Args:
            variables_of_interest (list or None): Lista de variables a incluir en las visualizaciones. Si es None, se utilizan todas las columnas del DataFrame.
            top_n_categories (int): N√∫mero de categor√≠as principales a mostrar en gr√°ficos de variables categ√≥ricas.

        Returns:
            None
        """
        #print("Aplicando plot_vars")
        plots_folder = os.path.join(self.output_folder, 'plots')
        os.makedirs(plots_folder, exist_ok=True)
        
        if variables_of_interest is None:
            variables_of_interest = data.columns

        numeric_variables = [var for var in variables_of_interest if pd.api.types.is_numeric_dtype(data[var])]
        categorical_variables = [var for var in variables_of_interest if pd.api.types.is_string_dtype(data[var])]
        
        # Plot de correlaci√≥n entre variables num√©ricas
        self.plot_correlation(data, numeric_variables, plots_folder)

        # Plot de matriz de correlaci√≥n de V de Cramer entre variables categ√≥ricas
        self.plot_cramers_heatmap(data, categorical_variables, plots_folder)
        
        for variable in variables_of_interest:
            if variable in data.columns:
                if pd.api.types.is_numeric_dtype(data[variable]):
                    self.plot_numeric_variable(data, variable, plots_folder)
                elif pd.api.types.is_string_dtype(data[variable]):
                    self.plot_categorical_variable(data, variable, top_n_categories, plots_folder)                     

    def plot_correlation(self, data, variables_of_interest, plots_folder):
        """
        Crea un gr√°fico de correlaci√≥n entre variables num√©ricas y lo guarda como un archivo PDF.

        Args:
            variables_of_interest (list): Lista de variables num√©ricas para las que se calcular√° y representar√° la correlaci√≥n.
            plots_folder (str): Directorio donde se guardar√° el archivo de gr√°ficos.

        Returns:
            None
        """
        correlation_matrix = data[variables_of_interest].corr()
        fig, ax = plt.subplots(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, ax=ax)
        plt.title('Correlaci√≥n entre variables num√©ricas')
        plot_filename = os.path.join(plots_folder, 'correlation.pdf')
        plt.savefig(plot_filename, format='pdf')
        plt.close()

    def plot_numeric_variable(self, data, variable, plots_folder):
        """
        Crea un gr√°fico de densidad para una variable num√©rica y lo guarda como un archivo PDF.

        Args:
            variable (str): Nombre de la variable num√©rica que se representar√°.
            plots_folder (str): Directorio donde se guardar√° el archivo de gr√°ficos.

        Returns:
            None
        """
        # Gr√°fico de densidad para variables num√©ricas usando Matplotlib
        plt.figure()
        plt.hist(data[variable], bins=30, color='blue', alpha=0.7)
        plt.xticks(rotation=45)
        plt.title(f'Distribuci√≥n de {variable}')
        plot_filename = os.path.join(plots_folder, f'{variable}.pdf')
        plt.savefig(plot_filename, format='pdf')
        plt.close()
        
    def plot_categorical_variable(self, data, variable, top_n_categories, plots_folder):
        """
        Crea un histograma de las categor√≠as m√°s frecuentes para una variable categ√≥rica y lo guarda como un archivo PDF.

        Args:
            variable (str): Nombre de la variable categ√≥rica que se representar√°.
            top_n_categories (int): N√∫mero de categor√≠as m√°s frecuentes a incluir en el histograma.
            plots_folder (str): Directorio donde se guardar√° el archivo de gr√°ficos.

        Returns:
            None
        """
        # Histograma de las top N categor√≠as usando Matplotlib
        top_categories = data[variable].value_counts().nlargest(top_n_categories)
        fig = plt.figure()
        top_categories.plot(kind='bar', color='green', alpha=0.7)
        plt.title(f'Top {top_n_categories} Categor√≠as de {variable}')
        plt.xticks(rotation=0)
        plot_filename = os.path.join(plots_folder, f'{variable}.pdf')
        plt.savefig(plot_filename, format='pdf')
        plt.close()


    # Define the Cramer's V function
    def cramers_V(self, var1, var2):
        """
        Calcula la medida de asociaci√≥n Cramer's V entre dos variables categ√≥ricas.

        Parameters:
        - var1: Serie de datos de la primera variable categ√≥rica.
        - var2: Serie de datos de la segunda variable categ√≥rica.

        Returns:
        - El valor de Cramer's V que indica la fortaleza de la asociaci√≥n entre las dos variables categ√≥ricas.
    
        Cramer's V es una estad√≠stica que mide la asociaci√≥n entre dos variables categ√≥ricas. Es una medida de la fuerza y direcci√≥n de la relaci√≥n entre las variables. Los valores de Cramer's V var√≠an entre 0 y 1, donde 0 indica ninguna asociaci√≥n y 1 indica una asociaci√≥n perfecta.

        La funci√≥n calcula Cramer's V utilizando una tabla de contingencia y el estad√≠stico Chi-cuadrado. Cuanto mayor sea el valor de Cramer's V, mayor ser√° la asociaci√≥n entre las dos variables categ√≥ricas.
        """
        crosstab = np.array(pd.crosstab(var1, var2, rownames=None, colnames=None))  # Cross table building
        stat = chi2_contingency(crosstab)[0]  # Keeping the test statistic of the Chi2 test
        obs = np.sum(crosstab)  # Number of observations
        mini = min(crosstab.shape) - 1  # Take the minimum value between the columns and the rows of the cross table
        return np.sqrt(stat / (obs * mini))
    
    # Define el m√©todo create_cramers_matrix 
    def create_cramers_matrix(self, data, categorical_variables):
        """
        Crea una matriz de Cramer's V que mide la asociaci√≥n entre variables categ√≥ricas.

        Parameters:
        - data: DataFrame que contiene los datos de las variables categ√≥ricas.
        - categorical_variables: Lista de nombres de las variables categ√≥ricas en el DataFrame.

        Returns:
        - Un DataFrame que representa la matriz de Cramer's V con valores redondeados.

        Este m√©todo calcula la asociaci√≥n entre todas las combinaciones posibles de variables categ√≥ricas en el DataFrame. Utiliza la medida de Cramer's V para cuantificar la fuerza de la asociaci√≥n. La matriz resultante muestra la asociaci√≥n entre cada par de variables categ√≥ricas.

        Cramer's V es una estad√≠stica que var√≠a entre 0 y 1, donde 0 indica ninguna asociaci√≥n y 1 indica una asociaci√≥n perfecta. Los valores de Cramer's V en la matriz est√°n redondeados a dos decimales para facilitar la interpretaci√≥n.
        """
        rows = []
        data_encoded = data.copy()
        for var1 in categorical_variables:
            col = []
            for var2 in categorical_variables:
                cramers = self.cramers_V(data_encoded[var1], data_encoded[var2])  # Cramer's V test
                col.append(round(cramers, 2))  # Keeping the rounded value of the Cramer's V
            rows.append(col)
        cramers_results = np.array(rows)
        df = pd.DataFrame(cramers_results, columns=categorical_variables, index=categorical_variables)
        return df

    
    def plot_cramers_heatmap(self, data, categorical_variables, plots_folder):
        """
        Crea un gr√°fico de la matriz de correlaci√≥n de V de Cramer entre variables categ√≥ricas y lo guarda como un archivo PDF.

        Args:
            categorical_variables (list): Lista de variables categ√≥ricas para las que se calcular√° y representar√° la matriz de correlaci√≥n de V de Cramer.
            plots_folder (str): Directorio donde se guardar√° el archivo de gr√°ficos.

        Returns:
            None
        """
        cramers_matrix = self.create_cramers_matrix(data, categorical_variables)
        mask = np.zeros_like(cramers_matrix, dtype=bool)
        mask[np.triu_indices_from(mask)] = True

        with sns.axes_style("white"):
            ax = sns.heatmap(cramers_matrix, mask=mask, vmin=0., vmax=1, square=True)

        plot_filename = os.path.join(plots_folder, 'cramers_matrix.pdf')
        plt.savefig(plot_filename, format='pdf')
        plt.close()
    
    # 4
    def clean_data(self, variables_of_interest=None, drop_duplicates=True, drop_na=False, impute_na=True, numeric_cols = ['age', 'height', 'weight'], use_categorical_vars = True):
        """
        Limpia y preprocesa los datos del Profiler.

        Args:
            variables_of_interest (list): Lista de variables de inter√©s a ser incluidas en el resultado limpio (por defecto, todas las variables).
            drop_duplicates (bool): Indica si se deben eliminar filas duplicadas (por defecto, True).
            drop_na (bool): Indica si se deben eliminar filas con valores nulos (por defecto, True).
            impute_na (bool): Indica si se deben imputar los valores nulos en las columnas (por defecto, False).
            numeric_cols (list): Lista de columnas num√©ricas a ser convertidas (por defecto, ['age', 'height', 'weight']).
            use_categorical_vars (bool): Indica si se deben utilizar las variables categ√≥ricas (por defecto, True).

        Returns:
            pandas.DataFrame
                Un DataFrame con la data procesada y limpia.
        """
        #print("Aplicando clean_data")
        def custom_impute_na(X):
            """
            Imputa valores nulos en un DataFrame dividiendo las columnas num√©ricas y categ√≥ricas,
            y reemplazando los valores nulos en las columnas num√©ricas con la media de la columna y
            los valores nulos en las columnas categ√≥ricas con la moda (valor m√°s frecuente).

            Args:
                X (DataFrame): El DataFrame que contiene las columnas a ser imputadas.

            Returns:
                DataFrame: El DataFrame resultante con los valores nulos imputados.
            """
            numeric_columns = X.select_dtypes(include=[np.number])
            categorical_columns = X.select_dtypes(exclude=[np.number])

            # Imputa valores nulos en columnas num√©ricas con la media
            numeric_columns = numeric_columns.fillna(numeric_columns.mean())

            # Imputa valores nulos en columnas categ√≥ricas con la moda (valor m√°s frecuente)
            categorical_columns = categorical_columns.fillna(categorical_columns.mode().iloc[0])

            return pd.concat([numeric_columns, categorical_columns], axis=1)
        
        def convert_numeric_columns(X, columns):
            """
            Convierte las columnas especificadas en el DataFrame a tipo num√©rico. Los valores no num√©ricos se
            convierten en NaN (Not-a-Number).

            Args:
                X (DataFrame): El DataFrame que contiene las columnas a convertir.
                columns (list): Una lista de nombres de columnas que se convertir√°n a tipo num√©rico.

            Returns:
                DataFrame: El DataFrame resultante con las columnas especificadas convertidas a tipo num√©rico.
            """
            for column in columns:
                X[column] = pd.to_numeric(X[column], errors='coerce')
            return X
        
        
        clean_data_folder = os.path.join(self.output_folder, 'clean_data')
        os.makedirs(clean_data_folder, exist_ok=True)

        data_to_clean = self.data.copy()
    
        # Detecta autom√°ticamente la columna con datos no at√≥micos
        non_atomic_column = self.detect_non_atomic_column(data_to_clean)

        if non_atomic_column:
            split_column_func = FunctionTransformer(lambda x: x.apply(self.split_column), validate=False)
            separated_data = split_column_func.transform(data_to_clean[non_atomic_column])

            # Verifica que todas las listas tengan la misma longitud
            if separated_data.apply(len).nunique() == 1:
                # Obtiene los elementos separados de non_atomic_column
                column_names = non_atomic_column.split('-')
                num_columns = len(column_names)

                # Verifica que la cantidad de elementos coincida con la longitud de las listas
                if num_columns == separated_data.apply(len).iloc[0]:
                    separated_data = pd.DataFrame(separated_data.tolist(), columns=column_names)
                    data_to_clean = pd.concat([data_to_clean, separated_data], axis=1)
                    data_to_clean.drop(columns=[non_atomic_column], inplace=True)
                else:
                    print("El n√∫mero de elementos separados no coincide con la longitud de las listas.")
            else:
                print("Las listas no tienen la misma longitud, no se pueden separar en columnas.")
        
        convert_numeric_func = FunctionTransformer(convert_numeric_columns, validate=False, kw_args={'columns': numeric_cols})
        data_to_clean = convert_numeric_func.transform(data_to_clean)
        
        
        if variables_of_interest is not None:
            data_to_clean = data_to_clean[variables_of_interest]

        if drop_duplicates:
            drop_duplicates_func = FunctionTransformer(lambda x: x.drop_duplicates(), validate=False)
            data_to_clean = drop_duplicates_func.transform(data_to_clean)

        if drop_na:
            drop_na_func = FunctionTransformer(lambda x: x.dropna(), validate=False)
            data_to_clean = drop_na_func.transform(data_to_clean)

        if impute_na:
            impute_na_func = FunctionTransformer(custom_impute_na, validate=False)
            data_to_clean = impute_na_func.transform(data_to_clean)
        
        if not use_categorical_vars:
            data_to_clean = data_to_clean.select_dtypes(include=[np.number])

        data_to_clean.to_csv(os.path.join(clean_data_folder, 'data.csv'), index=False)
        return data_to_clean
        
    def split_column(self, row):
        """
        Divide una columna no at√≥mica en elementos separados utilizando una expresi√≥n regular.

        Args:
            row (str): Valor de la columna no at√≥mica.

        Returns:
            list: Lista de elementos separados.
        """
        # Utiliza una expresi√≥n regular para buscar patrones de separadores
        return re.split(r'[*():?]', row)

    def detect_non_atomic_column(self, data):
        """
        Detecta autom√°ticamente una columna que contiene datos no at√≥micos.

        Args:
            data (DataFrame): Los datos a analizar.

        Returns:
            str: El nombre de la columna con datos no at√≥micos, o None si no se encuentra.
        """
        # Busca una columna que contenga caracteres especiales o separadores
        for column in data.columns:
            if any(char in column for char in ('-', ';', ':', '|')):
                return column
        return None
        
    def split_column(self, row):
        """
        Divide una columna no at√≥mica en elementos separados utilizando una expresi√≥n regular.

        Args:
            row (str): Valor de la columna no at√≥mica.

        Returns:
            list: Lista de elementos separados.
        """
        # Utiliza una expresi√≥n regular para buscar patrones de separadores
        return re.split(r'[*():?]', row)

    def detect_non_atomic_column(self, data):
        """
        Detecta autom√°ticamente una columna que contiene datos no at√≥micos.

        Args:
            data (DataFrame): Los datos a analizar.

        Returns:
            str: El nombre de la columna con datos no at√≥micos, o None si no se encuentra.
        """
        # Busca una columna que contenga caracteres especiales o separadores
        for column in data.columns:
            if any(char in column for char in ('-', ';', ':', '|')):
                return column
        return None
    
    def _clean_data_wrapper(self, X, y=None):
        """
        Devuelve los datos limpios independientemente de la entrada.

        Este m√©todo es un envoltorio que ignora cualquier entrada y proporciona
        una interfaz consistente para su uso en una pipeline de scikit-learn o
        transformadores similares que esperan una funci√≥n con entradas X e y.

        Par√°metros:
        ----------
        X : Este par√°metro es ignorado.
        y : Ignorado, opcional (default=None)
            Este par√°metro es ignorado.

        Retorna:
        -------
        pandas.DataFrame
            Un DataFrame que contiene los datos limpios obtenidos a trav√©s del
            m√©todo clean_data de la misma clase.
        """
        # Esta funci√≥n simplemente devuelve la data limpia.
        return self.clean_data(use_categorical_vars=True)

    def _clean_data_wrapper_num(self, X, y=None):
        """
        Devuelve los datos limpios independientemente de la entrada, considerando 
        solo las columnas num√©ricas.

        Este m√©todo es un envoltorio que ignora cualquier entrada y proporciona
        una interfaz consistente para su uso en una pipeline de scikit-learn o
        transformadores similares que esperan una funci√≥n con entradas X e y.

        Par√°metros:
        ----------
        X : Este par√°metro es ignorado.
        y : Ignorado, opcional (default=None)
            Este par√°metro es ignorado.

        Retorna:
        -------
        pandas.DataFrame
            Un DataFrame que contiene los datos limpios obtenidos a trav√©s del
            m√©todo clean_data de la misma clase, considerando s√≥lo las columnas num√©ricas.
        """
        # Esta funci√≥n simplemente devuelve la data limpia.
        return self.clean_data(use_categorical_vars=False)
    
    
    # 5
    def scale(self, categorical_transformer=OneHotEncoder(sparse_output=False), use_categorical_vars = True):
        """
        Escala y transforma los datos num√©ricos y categ√≥ricos del conjunto de datos.

        Este m√©todo aplica una una codificaci√≥n a las variables categ√≥ricas y luego transformaci√≥n 
        logar√≠tmica y una normalizaci√≥n Min-Max a todas las variables. La salida es un DataFrame 
        preparado para el modelado de machine learning.

        Par√°metros:
        ----------
        categorical_transformer : transformer, opcional (default=OneHotEncoder(sparse_output=False))
            Un transformador que sigue la API de scikit-learn (debe tener los m√©todos 'fit' y 'transform').
            Se utiliza para transformar las variables categ√≥ricas. Por defecto, se utiliza OneHotEncoder.
        use_categorical_vars (bool): Indica si se deben utilizar las variables categ√≥ricas (por defecto, True).

        Retorna:
        -------
        pandas.DataFrame
            Un DataFrame con las caracter√≠sticas escaladas y las variables categ√≥ricas codificadas, si es que son 
            consideradas por use_categorical_vars.

        Levanta:
        ------
        ValueError
            Si el 'categorical_transformer' proporcionado no cumple con la API de scikit-learn (m√©todos 'fit' y 'transform').
        """
        #print("Aplicando scale")
        if not (hasattr(categorical_transformer, 'fit') and hasattr(categorical_transformer, 'transform')):
            raise ValueError("El algoritmo de transformaci√≥n de variables categ√≥ricas proporcionado no es compatible con sklearn")
        
        scale_folder = os.path.join(self.output_folder, 'scale')
        os.makedirs(scale_folder, exist_ok=True)

        total_clean_data = self.clean_data(use_categorical_vars=use_categorical_vars)

        numerical_columns = total_clean_data.select_dtypes(include=[np.number]).columns.tolist()

        
        # Definimos el transformador para los datos num√©ricos
        numeric_transformer_pipeline = Pipeline(steps=[
            ('log1p', FunctionTransformer(np.log1p, validate=False)),
            ('scaler', MinMaxScaler())
        ])

        clean_data_transformer = FunctionTransformer(func=self._clean_data_wrapper, validate=False)
        if use_categorical_vars:
            categorical_columns = total_clean_data.select_dtypes(exclude=[np.number]).columns.tolist()
            # Definimos el transformador para los datos categ√≥ricos
            categorical_transformer_pipeline = Pipeline(steps=[
                ('cat', categorical_transformer),
                ('log1p', FunctionTransformer(np.log1p, validate=False)),
                ('scaler', MinMaxScaler())
            ])

            # Creamos el ColumnTransformer
            preprocessor = ColumnTransformer(
                transformers=[
                    ('cat', categorical_transformer_pipeline, categorical_columns), 
                    ('num', numeric_transformer_pipeline, numerical_columns),
                ]
            )
        else:
            #clean_data_transformer = FunctionTransformer(func=self._clean_data_wrapper_num, validate=False)
            # Creamos el ColumnTransformer
            preprocessor = ColumnTransformer(
                transformers=[
                    ('num', numeric_transformer_pipeline, numerical_columns),
                ]
            )


        # Creamos el Pipeline completo
        full_pipeline = Pipeline(steps=[
            ('clean_data', clean_data_transformer),  # Agrega el limpiador de datos al principio del pipeline
            ('preprocess', preprocessor)  # El resto del preprocesamiento como estaba antes
        ])



        # Obtenemos los nombres de las nuevas columnas generadas por el OneHotEncoder
        # Suponiendo que 'cat' es el nombre del paso en su pipeline de transformaci√≥n categ√≥rica
        if use_categorical_vars:
            # Ajustamos el ColumnTransformer a los datos
            #preprocessor.fit(clean_data)
            #full_pipeline.fit(self.data)

            # Transformamos los datos con el ColumnTransformer
            #scaled_data = preprocessor.transform(clean_data)
            #scaled_data = full_pipeline.transform(self.data)
            scaled_data = full_pipeline.fit_transform(total_clean_data)
            categorical_feature_names = preprocessor.named_transformers_['cat']['cat'].get_feature_names_out(categorical_columns)
            # Combinamos los nombres de las columnas num√©ricas y categ√≥ricas
            new_column_names = np.concatenate((categorical_feature_names, numerical_columns))
        else:
            # Ajustamos el ColumnTransformer a los datos
            #preprocessor.fit(clean_data)
            #full_pipeline.fit(self.data.select_dtypes(include=[np.number]))

            # Transformamos los datos con el ColumnTransformer
            #scaled_data = preprocessor.transform(clean_data)
            #scaled_data = full_pipeline.transform(self.data.select_dtypes(include=[np.number]))
            scaled_data = full_pipeline.fit_transform(total_clean_data)
            new_column_names = numerical_columns


        # Creamos un DataFrame con los datos transformados y los nuevos nombres de columnas
        df_processed = pd.DataFrame(scaled_data, columns=new_column_names)

        df_processed.to_csv(f'{scale_folder}/scaled_features.csv', index=False)

        return df_processed
    
    def _scale_wrapper(self, X, y=None):
        """
        Devuelve los datos escalados independientemente de la entrada X.

        Este m√©todo es un envoltorio que ignora la entrada X y utiliza el m√©todo scale
        de la clase para devolver datos ya escalados. Est√° dise√±ado para ser compatible
        con las expectativas de entrada/salida de los componentes de scikit-learn como
        FunctionTransformer.

        Par√°metros:
        ----------
        X : Ignorado
            Este par√°metro es ignorado y no influye en el resultado.
        y : Ignorado, opcional (default=None)
            Este par√°metro es ignorado y no se utiliza en el proceso.

        Retorna:
        -------
        pandas.DataFrame
            Retorna un DataFrame de los datos despu√©s de aplicar la funci√≥n¬†scale.
        """
        # Esta funci√≥n devuelve la data escalada.
        
        return self.scale(use_categorical_vars=False)
    
    def _scale_wrapper_num(self, X, y=None):
        """
        Devuelve los datos escalados independientemente de la entrada X.

        Este m√©todo es un envoltorio que ignora la entrada X y utiliza el m√©todo scale
        de la clase para devolver datos ya escalados, considerando solo las columnas 
        num√©ricas. Est√° dise√±ado para ser compatible con las expectativas de 
        entrada/salida de los componentes de scikit-learn como FunctionTransformer.

        Par√°metros:
        ----------
        X : Ignorado
            Este par√°metro es ignorado y no influye en el resultado.
        y : Ignorado, opcional (default=None)
            Este par√°metro es ignorado y no se utiliza en el proceso.

        Retorna:
        -------
        pandas.DataFrame
            Retorna un DataFrame de los datos despu√©s de aplicar la funci√≥n¬†scale, considerando
            solo las columnas num√©ricas.
        """
        # Esta funci√≥n devuelve la data escalada.
        
        return self.scale(use_categorical_vars=False)
    # 6
    def make_clusters(self, clustering_algorithm = KMeans(n_init='auto'), n_clusters_range=(2, 12), plot_elbow=True, use_categorical_vars=False):
        """
        Realiza el an√°lisis de clustering en los datos transformados por la clase. Esta funci√≥n admite cualquier
        algoritmo de clustering de scikit-learn que sea una instancia de ClusterMixin. Tambi√©n puede determinar
        el n√∫mero √≥ptimo de cl√∫steres si se solicita.

        Par√°metros:
        clustering_algorithm : object, opcional (default=KMeans(n_init='auto'))
            El algoritmo de clustering para usar. Debe ser una instancia de ClusterMixin de scikit-learn.
            Por defecto, utiliza KMeans con una inicializaci√≥n autom√°tica.
        
        n_clusters_range : tuple, opcional (default=(2, 12))
            Un rango de n√∫meros de clusters a considerar si se est√° usando el m√©todo de codo para determinar
            el n√∫mero √≥ptimo de clusters. Esto ser√° pasado al KElbowVisualizer.
        
        plot_elbow : bool, opcional (default=True)
            Si es True, la funci√≥n buscar√° el n√∫mero √≥ptimo de cl√∫steres utilizando el m√©todo de codo
            y lo graficar√°. Adem√°s, ajustar√° el algoritmo de clustering proporcionado con este n√∫mero √≥ptimo.
        use_categorical_vars (bool): Indica si se deben utilizar las variables categ√≥ricas (por defecto, False).

        Retorna:
        cluster_data : DataFrame
            Un DataFrame con los datos proyectados en 2 dimensiones principales y la asignaci√≥n de cl√∫steres.

        Levanta:
        ValueError
            Si el algoritmo de clustering proporcionado no es una instancia de ClusterMixin de scikit-learn.
        """
        #print("Aplicando make_clusters")
        # Comprobar si el algoritmo de clustering es una instancia de ClusterMixin para asegurar compatibilidad con sklearn
        if not isinstance(clustering_algorithm, ClusterMixin):
            raise ValueError("El algoritmo de clustering proporcionado no es compatible con sklearn")
        
        # Crear la carpeta para guardar los resultados de los clusters
        clusters_folder = os.path.join(self.output_folder, 'clusters')
        os.makedirs(clusters_folder, exist_ok=True)
        
        # Cargar y escalar los datos
        transformed_data = self.scale(use_categorical_vars=use_categorical_vars) 

        # Inicializar el n√∫mero √≥ptimo de cl√∫steres con None
        optimal_n_clusters = None
        
        # Determinar el n√∫mero √≥ptimo de cl√∫steres si es necesario y el algoritmo lo soporta
        if plot_elbow:
            elbow_model = KElbowVisualizer(clustering_algorithm, k=n_clusters_range)
            elbow_model.fit(transformed_data)
            elbow_model.show(outpath=os.path.join(clusters_folder, 'elbow_plot.png'))
            optimal_n_clusters = elbow_model.elbow_value_
            # Establecer el n√∫mero √≥ptimo de cl√∫steres en el algoritmo
            clustering_algorithm.set_params(n_clusters=optimal_n_clusters)
        
        if use_categorical_vars:
            scale_transformer = FunctionTransformer(func=self._scale_wrapper, validate=False)
        else:
            scale_transformer = FunctionTransformer(func=self._scale_wrapper_num, validate=False)
         # Creamos el Pipeline completo
        full_pipeline = Pipeline(steps=[
            ('scale', scale_transformer), 
            ('clustering', clustering_algorithm) 
        ])
        # Ajustar el pipeline a los datos ya escalados
        #clusters = pipeline.fit_transform(transformed_data)
        labels = full_pipeline.fit_predict(transformed_data)

        # Proyectar los datos a 2 dimensiones para visualizaci√≥n
        decomposition = PCA(n_components=2)

        reduced_data = decomposition.fit_transform(transformed_data)
        
        # Crear un DataFrame para la visualizaci√≥n y guardar los resultados
        cluster_data = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])
        cluster_data['Cluster'] = labels

        # Crear un DataFrame y guardar los resultados
        cluster_data_with_all_cols = pd.DataFrame(transformed_data)
        cluster_data_with_all_cols['Cluster'] = labels
        cluster_data_with_all_cols.to_csv(os.path.join(clusters_folder, 'cluster_data.csv'), index=False)
        
        # Graficar los cl√∫steres
        plt.figure(figsize=(10, 8))
        plt.scatter(cluster_data['PC1'], cluster_data['PC2'], c=cluster_data['Cluster'], cmap='viridis', marker='o')
        plt.title('Visualizaci√≥n de Clusters')
        plt.xlabel('PC 1')
        plt.ylabel('PC 2')
        # Creamos una lista de colores basados en los clusters √∫nicos en 'labels'
        unique_labels = np.unique(labels)
        colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels)))

        # Creamos una lista de leyendas, una para cada cluster
        legend_handles = [Line2D([0], [0], marker='o', color='w', label=f'Cluster {label}',
                                markerfacecolor=color, markersize=10) for label, color in zip(unique_labels, colors)]

        # A√±adimos la leyenda al gr√°fico con una ubicaci√≥n espec√≠fica fuera del √°rea de trazado
        plt.legend(handles=legend_handles, title='Clusters', loc='upper left', bbox_to_anchor=(1,1))

        # Ajustar el layout para dar espacio a la leyenda fuera del gr√°fico
        plt.tight_layout()
        plt.savefig(os.path.join(clusters_folder, 'clusters_visualization.png'))
        plt.show()

        return cluster_data, cluster_data_with_all_cols
        
    # 7
    def detect_anomalies(self, algorithm = IsolationForest(n_jobs=-1)):
        """
        Detecta anomal√≠as en el conjunto de datos utilizando el algoritmo de detecci√≥n de 
        anomal√≠as de scikit-learn especificado.

        Par√°metros:
        algorithm : scikit-learn BaseEstimator (por defecto=IsolationForest(n_jobs=-1))
            El algoritmo de detecci√≥n de anomal√≠as a utilizar. Debe ser un estimador de 
            scikit-learn que soporte el m√©todo fit_predict..
        
        Retorna:
        numpy.ndarray
            Las etiquetas predichas para cada instancia en el conjunto de datos, donde -1 indica 
            una anomal√≠a y 1 indica un valor normal.

        Levanta:
        ValueError: Si el algoritmo proporcionado no es un estimador de scikit-learn v√°lido.
        """
        #print("Aplicando detect_anomalies")
        # Verificar que el algoritmo pertenezca a sklearn
        if not issubclass(type(algorithm), BaseEstimator):
            raise ValueError("El algoritmo proporcionado no es un estimador de scikit-learn v√°lido.")

        # Crear carpeta para anomalies
        path_anomalies = os.path.join(self.output_folder, 'anomalies')
        os.makedirs(path_anomalies, exist_ok=True)

        # Preprocesar y escalar los datos
        preprocessed_scaled_data_original = self.scale()
        preprocessed_scaled_data = preprocessed_scaled_data_original.copy()

        # Crear el pipeline con solo el algoritmo de detecci√≥n de anomal√≠as
        scale_transformer = FunctionTransformer(func=self._scale_wrapper, validate=False)

        pipeline = Pipeline(steps=[
            ('scale', scale_transformer), 
            ('anomaly_detector', algorithm) 
        ])

        scores =  pipeline.fit_predict(preprocessed_scaled_data)

        # Apply PCA
        decomposition = PCA(n_components=2)
        data_2d = decomposition.fit_transform(preprocessed_scaled_data)

        # Usar colores espec√≠ficos para -1 y 1.
        colors = np.where(scores == -1, 'red', 'blue')  # 'red' para -1, 'blue' para 1.

        # Plotting the results
        plt.scatter(data_2d[:, 0], data_2d[:, 1], c=colors)

        # Crear handles personalizados
        legend_elements = [Line2D([0], [0], marker='o', color='w', label='Data Anomala',
                                markerfacecolor='red', markersize=10),
                        Line2D([0], [0], marker='o', color='w', label='Data Normal',
                                markerfacecolor='blue', markersize=10)]

        # Crear la leyenda con los handles personalizados
        plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left')

        plt.title('Detecci√≥n de Anomal√≠as')
        plt.xlabel('PC 1')
        plt.ylabel('PC 2')

        # Asegurarse de que el layout se ajuste para no cortar la leyenda
        plt.tight_layout()
        plt.savefig(os.path.join(path_anomalies, 'anomaly_detection.png'))


        # Agregar etiquetas de anomal√≠as al DataFrame y guardar
        data_with_anomalies = preprocessed_scaled_data
        data_with_anomalies['anomaly_label'] = scores
        data_with_anomalies.to_csv(os.path.join(path_anomalies, 'data_anomalies.csv'), index=False)

        return scores

    
    #8
    def profile(self, variables_of_interest_summarize=None, criteria_for_outliers=None, 
                variables_of_interest_plot=None, top_n_categories=5, 
                use_categorical_vars_clean=True,
                variables_of_interest_clean_data=None, 
                drop_duplicates=True, 
                drop_na=False, 
                impute_na=True, 
                numeric_cols = ['age', 'height', 'weight'], 
                scaling_technique = OneHotEncoder(sparse_output=False), 
                scale_categorical_vars = True,
                clustering_algorithm = KMeans(n_init='auto'), 
                cluster_categorical_vars = False,
                n_clusters_range = (2, 12), 
                plot_elbow = True, 
                anomaly_detection_algorithm=IsolationForest(n_jobs=-1)):
        """
        Realiza un an√°lisis de perfil completo del conjunto de datos, incluyendo resumen, 
        detecci√≥n de outliers, visualizaciones, limpieza de datos, escalado, clustering y 
        detecci√≥n de anomal√≠as.

        Par√°metros:
        ----------
        variables_of_interest_summarize : list of str, opcional
            Variables a resumir en el m√©todo summarize.
        criteria_for_outliers : dict, opcional
            Criterios espec√≠ficos para identificar outliers en el m√©todo summarize.
        variables_of_interest_plot : list of str, opcional
            Variables a visualizar en el m√©todo plot_vars.
        top_n_categories : int, default=5
            Top N categor√≠as a mostrar en las visualizaciones de variables categ√≥ricas.
        use_categorical_vars_clean : Indica si se deben utilizar las variables categ√≥ricas en el clean_data (por defecto, True).
        variables_of_interest_clean_data : list of str, opcional
            Variables a incluir en el proceso de limpieza de datos.
        drop_duplicates : bool, default=True
            Si se deben eliminar duplicados en el m√©todo de limpieza de datos.
        drop_na : bool, default=True
            Si se deben eliminar las filas con valores NA en el m√©todo de limpieza de datos.
        impute_na : bool, default=False
            Si se deben imputar los valores NA en el m√©todo de limpieza de datos.
        numeric_cols : list of str, default=['age', 'height', 'weight']
            Columnas num√©ricas a considerar en el proceso de escalado.
        scaling_technique : transformer, default=OneHotEncoder(sparse_output=False)
            T√©cnica de escalado para aplicar a los datos antes de la clusterizaci√≥n.
        scale_categorical_vars : Indica si se deben utilizar las variables categ√≥ricas en el scale (por defecto, True). 
        clustering_algorithm : clustering instance, default=KMeans(n_init='auto')
            Algoritmo de clustering para usar en la generaci√≥n de cl√∫steres.
        cluster_categorical_vars : Indica si se deben utilizar las variables categ√≥ricas en el make_clusters (por defecto, False). 
        n_clusters_range : tuple, default=(2, 12)
            Rango de n√∫mero de cl√∫steres a considerar para el m√©todo 'make_clusters'.
        plot_elbow : bool, default=True
            Si se debe generar o no el gr√°fico del codo para ayudar a determinar el n√∫mero √≥ptimo de cl√∫steres.
        anomaly_detection_algorithm : anomaly detection instance, default=IsolationForest(n_jobs=-1)
            Algoritmo para la detecci√≥n de anomal√≠as en el conjunto de datos.

        Retorna:
        -------
        tuple
            Una tupla que contiene los resultados de los siguientes procesos:
            - Resumen de estad√≠sticas descriptivas
            - Visualizaci√≥n de variables
            - Datos limpios
            - Datos escalados
            - Cl√∫steres generados
            - Anomal√≠as¬†detectadas
        """
        #print("Aplicando profile")
        # Ejecutar el m√©todo summarize
        summ = self.summarize(variables_of_interest=variables_of_interest_summarize, criteria_for_outliers=criteria_for_outliers)
        
        # Limpiar los datos
        clean_data = self.clean_data(variables_of_interest=variables_of_interest_clean_data, drop_duplicates=drop_duplicates, 
                                     drop_na=drop_na, impute_na=impute_na, numeric_cols=numeric_cols, use_categorical_vars = use_categorical_vars_clean)
        
        # Ejecutar el m√©todo plot_vars
        plt_vars = self.plot_vars(data=clean_data ,variables_of_interest=variables_of_interest_plot, top_n_categories=top_n_categories)
        
        # Escalar los datos
        scaled_data = self.scale(categorical_transformer=scaling_technique, use_categorical_vars = scale_categorical_vars)
        
        # Ejecutar clustering
        clusters = self.make_clusters(clustering_algorithm=clustering_algorithm, 
                                      n_clusters_range=n_clusters_range, 
                                      plot_elbow=plot_elbow, 
                                      use_categorical_vars=cluster_categorical_vars)

        # Detectar anomal√≠as
        anomalies = self.detect_anomalies(anomaly_detection_algorithm)

        return summ, plt_vars, clean_data, scaled_data, clusters, anomalies



    # 9
    def clearGarbage(self):
        """
        Elimina todas las carpetas y archivos creados por la clase Profiler. 
        """
        print("Aplicando clearGarbage")
        if os.path.exists(self.output_folder):
            shutil.rmtree(self.output_folder)
    
```
