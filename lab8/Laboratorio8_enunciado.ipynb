````python
df_players.drop(['National_Position', 'Nationality', 'Name'], axis=1, inplace=True)

categorical_columns = ['Club_Position', 'Preffered_Foot', 'Work_Rate']
numeric_columns = [
    'Height', 'Weight', 'Age', 'Weak_foot', 'Skill_Moves', 'Ball_Control', 'Dribbling', 'Marking', 'Sliding_Tackle',
    'Standing_Tackle', 'Aggression', 'Reactions', 'Interceptions', 'Vision', 'Composure', 'Crossing', 'Short_Pass',
    'Long_Pass', 'Acceleration', 'Speed', 'Stamina', 'Strength', 'Balance', 'Agility', 'Jumping', 'Heading',
    'Shot_Power', 'Finishing', 'Long_Shots', 'Curve', 'Freekick_Accuracy', 'Penalties', 'Volleys'
]


ct = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(), categorical_columns),  
    ('scaler', MinMaxScaler(), numeric_columns)  
], remainder='passthrough')
````

````python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Define el pipeline para SVC
svc_pipeline = Pipeline([
    ('preprocessor', ct),  
    ('classifier', SVC(kernel='linear'))  
])

# Define el pipeline para K-Nearest Neighbors 
knn_pipeline = Pipeline([
    ('preprocessor', ct),  
    ('classifier', KNeighborsClassifier(n_neighbors=5))  
])

# Define el pipeline para RandomForest
rf_pipeline = Pipeline([
    ('preprocessor', ct),  
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=314159))  
])
````

````python
X = df_players.drop('Seleccionado', axis = 1)
y = df_players['Seleccionado']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=314159, stratify=y)
````

````python
# Entrena el pipeline para SVC lineal
svc_pipeline.fit(X_train, y_train)

# Entrena el pipeline para K-Nearest Neighbors
knn_pipeline.fit(X_train, y_train)

# Entrena el pipeline para RandomForestClassifier
rf_pipeline.fit(X_train, y_train)
````

````python
# Hacer predicciones en los datos de prueba.
y_pred_svc = svc_pipeline.predict(X_test)
y_pred_knn = knn_pipeline.predict(X_test)
y_pred_rf = rf_pipeline.predict(X_test)

# Calcular el informe de clasificación para cada modelo.
report_svc = classification_report(y_test, y_pred_svc)
report_knn = classification_report(y_test, y_pred_knn)
report_rf = classification_report(y_test, y_pred_rf)

# Imprimir los informes de clasificación.
print("Classification Report for SVC Pipeline:")
print(report_svc)

print("Classification Report for K-Nearest Neighbors Pipeline:")
print(report_knn)

print("Classification Report for Random Forest Pipeline:")
print(report_rf)
````
