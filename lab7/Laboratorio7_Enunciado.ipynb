## 1.2 Creaci√≥n de nuevas Caracteristicas [2 Puntos] 

Como ya se les comento, Mr. Lepin esta interesado en obtener las caracter√≠sticas **LRMFP**, para esto les se√±ala que estas caracter√≠sticas se construyen en base a las siguientes definiciones:

- **Length (L)**: Intervalo de tiempo, en d√≠as, entre la primera y la √∫ltima visita del cliente. Mientras mas grande sea el valor, mas fiel es el cliente.

- **Recency (R)**: Indica hace cuanto tiempo el cliente realizo su ultima compra. Notar que para este caso, mientras mas grande es el valor, menos interes posee el usuario para repetir una compra en uno de los locales.

- **Monetary (M)**: El t√©rmino "monetario" se refiere a la cantidad media de dinero gastada por cada visita del cliente durante el per√≠odo de observaci√≥n y refleja la contribuci√≥n del cliente a los ingresos de la empresa.

- **Frequency (F)**: Se refiere al n√∫mero total de visitas del cliente durante el periodo de observaci√≥n. Cuanto mayor sea la frecuencia, mayor ser√° la fidelidad del cliente. 

- **Periodicity (P)**: Representa si los clientes visitan las tiendas con regularidad.

$$Periodicity(n)=std(IVT_1, ..., IVT_n)$$

&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Donde $IVT$ denota el tiempo entre visitas y n representa el n√∫mero de valores de tiempo entre visitas de un cliente.
 

$$IVT_i=date\_diff(t_{i+1},t)$$

En base a las definiciones se√±aladas, dise√±e una funci√≥n que permita obtener las caracter√≠sticas **LRMFP** recibiendo un DataFrame como entrada. Para esto, no estar√° permitido el uso de iteradores, utilice todas las herramientas que les ofrece `pandas` para realizar esto.

Una referencia que le puede ser √∫til es el [documento original](https://www.researchgate.net/publication/315979555_LRFMP_model_for_customer_segmentation_in_the_grocery_retail_industry_a_case_study) en donde se propone este m√©todo.

```python
def custom_features(dataframe_in):
    resultados = []
    
    # Crear una lista de Customer IDs √∫nicos
    customer_ids = dataframe_in['Customer ID'].unique().astype(int)
    customer_ids.sort()
    
    #Calculo de length
    length = (dataframe_in.groupby('Customer ID')['InvoiceDate'].max() - dataframe_in.groupby('Customer ID')['InvoiceDate'].min()).dt.days
    
    #Calculo de recency
    fecha_mas_reciente = dataframe_in['InvoiceDate'].max() + pd.DateOffset(days=1)
    recency = (fecha_mas_reciente - dataframe_in.groupby('Customer ID')['InvoiceDate'].max()).dt.days
    
    #Calculo de frequency
    frequency = dataframe_in.groupby('Customer ID')['InvoiceDate'].nunique()
    
    #Calculo de monetary
    monetary = dataframe_in.groupby(['Customer ID', 'InvoiceDate']).apply(lambda x: (x['Price'] * x['Quantity']).sum()).groupby('Customer ID').mean().round(2)

    time_between_visits = dataframe_in.groupby('Customer ID')['InvoiceDate'].diff()
    periodicity = time_between_visits.dt.total_seconds().groupby(dataframe_in['Customer ID']).std()

    # Identify customers with only one visit
    single_or_two_visit_customers = frequency[(frequency == 1) | (frequency == 2)].index

    # Replace NaN values with 0 for these customers
    periodicity.loc[single_or_two_visit_customers] = 0

    dataframe_out = pd.DataFrame({'Length': length,
                                  'Recency': recency,
                                  'Frequency': frequency,
                                  'Monetary': monetary,
                                  'Periodicity': periodicity})
    return(dataframe_out)
    
```

```python
df_out = custom_features(df_retail)
df_out
```

### 1.3.1 Estandarizar Caracteristicas [0.5 puntos]

Construya una clase llamada ``MinMax()`` utilizando ``BaseEstimator`` y ``TransformerMixin`` para realizar una transformaci√≥n de cada una de las columnas de un DataFrame utilizando ``ColumnTransformer()`` m√°s tarde (tome como referencia el siguiente [enlace](https://sklearn-template.readthedocs.io/en/latest/user_guide.html#transformer)).


 Para esto considere que Min-Max escaler queda dada por la ecuaci√≥n:

$$MinMax = \dfrac{x-min(x)}{max(x) - min(x)}$$

Con esto buscamos que los valores que componen a las columnas se muevan en el rango de valores $[0, 1]$.

**Respuesta:**

```python
class MinMax(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        min_val = X.min()
        max_val = X.max()
        scaled_data = (X - min_val) / (max_val - min_val)
        return scaled_data
```

### 1.3.2 `T-SNE` Pipeline [1.0 puntos]

Para comenzar introduci√©ndose en el uso de pipeline, decide probar realizando un pipeline enfocado en la reducci√≥n de dimensionalidad y as√≠ hacer no decepcionar a Mr. Lepin con la clusterizaci√≥n del modelo. 

Configure un pipeline utilizando el algoritmo `T-SNE` sobre los datos **LRMFP**, donde, para la realizaci√≥n del pipeline considera los siguientes pasos:

1. Como primer paso obtenga las caracter√≠sticas **LRMFP** desde el DataFrame ``df_retail_II_cleaned.pickle`` utilizando la funci√≥n ``custom_features`` creada anteriormente, junto a ``FunctionTransformer()``. Considere esto como el primer paso de su pipeline.
2. En segundo lugar usando ``ColumnTransformer()`` aplique el MinxMax scaler creado por usted sobre todas las columnas generadas en el paso anterior. 
3. Finalmente, aplique un √∫ltimo paso donde obtiene las 2 componentes m√°s relevantes utilizando el algoritmo `T-SNE` de sckit-learn.

Tras aplicar las transformaciones sobre el dataset **LRMFP**, gr√°fique las componentes obtenidas en la reducci√≥n de dimensionalidad.

**Respuesta:**

```python
# 1.
fn_custom_features = FunctionTransformer(func=custom_features)

# 2. 
column_transformer = ColumnTransformer(
    transformers=[
        ('minmax', MinMax(), ['Length', 'Recency', 'Frequency', 'Monetary', 'Periodicity']),  
    ],
    remainder='passthrough' 
)

# 3.
tsne = TSNE(n_components=2, random_state=314159)


pipeline = Pipeline([('df_out', fn_custom_features),
                     ('MinMax', column_transformer),
                     ('TSNE', tsne)
                     ])

lrmfp = pipeline.fit_transform(df_retail)
```

```python
lrmfp
```

```python
# Obtener las dos componentes principales
tsne_components = lrmfp[:, :2]

# Crear un DataFrame para facilitar la visualizaci√≥n
tsne_df = pd.DataFrame(tsne_components, columns=['Componente 1', 'Componente 2'])

# Crea un gr√°fico de dispersi√≥n con Plotly Express
fig = px.scatter(tsne_df, x='Componente 1', y='Componente 2', title='T-SNE Components')
fig.show()
```

### 1.3.3 Clustering

#### 1.3.3.1 M√©todo del Codo [1 puntos]

Utilizando la clase creada para escalamiento, aplique el m√©todo del codo para visualizar cual es el n√∫mero de clusters que mejor se ajustan a los datos. Realice esto utilizando el algoritmo K-means dentro de un pipeline para un $k \in [1,20]$, donde k representa el n√∫mero de clusters del k-means. Para la realizaci√≥n de esta secci√≥n y la pr√≥xima (1.3.3.2), considere los mismos pasos utilizados para el t-sne, pero **permutando el algoritmo de reducci√≥n de dimensionalidad por k-means.**

A trav√©s del grafico obtenido, comente y justifique que valor de k escoger√≠a para realizar el k-means.

**Respuesta:**

Al examinar el gr√°fico de la inercia en funci√≥n de k, se observa que la inercia comienza a aplanarse alrededor de k=3. Esto indica que, a partir de k=3, la reducci√≥n en la inercia no es tan significativa en comparaci√≥n con los valores anteriores. Por lo tanto, se selecciona k=3 como el valor √≥ptimo para realizar el algoritmo K-means. Esta elecci√≥n ofrece un equilibrio adecuado entre la segmentaci√≥n de los datos en grupos distintos y la minimizaci√≥n de la varianza dentro de los cl√∫steres. Adem√°s, un valor de k m√°s alto podr√≠a dar lugar a una subdivisi√≥n excesiva de los datos, lo que complicar√≠a la interpretaci√≥n y el uso de los resultados. Por otro lado, el cambio en la pendiente de entre 1 - 2 y 2 - 3 es significativo, por lo que podr√≠amos considerar el 2 como punto √≥ptimo, pero la pendiente si la prolongamos sigue siendo muy empinada por lo que finalmente nos quedamos con el punto 3 donde se ve que la pendiente de 3 a 4 si la prolongamos ya es m√°s plana. Por lo tanto, k=3 se presenta como la elecci√≥n m√°s razonable y justificada en este contexto. Finalmente, para confirmar, se utiliza la librer√≠a kneed y se busca el √≥ptimo de manera heur√≠stica mediante KneeLocator.

```python
# Crear una lista para almacenar los valores de inercia (suma de cuadrados dentro de los cl√∫steres)
inertia_values = []

# Rango de k (n√∫mero de clusters)
k_range = range(1, 10)

for k in k_range:
    # Crear un pipeline similar al anterior, pero usando K-means con k clusters
    kmeans = KMeans(n_clusters=k, random_state=314159)
    
    pipeline = Pipeline([
        ('df_out', fn_custom_features),
        ('MinMax', column_transformer),
        ('KMeans', kmeans)
    ])
    
    # Ajustar el modelo K-means y calcular la inercia
    pipeline.fit(df_retail)
    inertia_values.append(pipeline.named_steps['KMeans'].inertia_)

# Crear un gr√°fico de dispersi√≥n con Plotly Express
fig = px.scatter(x=list(k_range), y=inertia_values, title='M√©todo del Codo para determinar el n√∫mero √≥ptimo de clusters')
fig.update_traces(mode='lines+markers')
fig.update_xaxes(title='N√∫mero de Clusters (k)')
fig.update_yaxes(title='Inercia')
fig.show()


# Detecci√≥n autom√°tica del n√∫mero √≥ptimo de clusters (punto del codo)
from kneed import KneeLocator

kneedle = KneeLocator(list(k_range), inertia_values, curve='convex', direction='decreasing')
optimal_k = kneedle.elbow

print(f'El n√∫mero √≥ptimo de clusters es: {optimal_k}')
```
