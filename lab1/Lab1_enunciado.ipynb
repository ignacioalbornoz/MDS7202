1. Git es un sistema de control de versiones que permite a desarrolladores, o personas que estén trabajando colaborativamente en ciertos proyectos, a tener un control sobre los archivos generados a lo largo del tiempo. En este sentido, es mayoritariamente utilizado por desarrolladores para tener un seguimiento de los cambios realizados en el código fuente de un proyecto. De esta manera resulta más fácil el trabajar colaborativamente, permitiendo que se trabaje en ramas paralelas, ver los cambios de los demás miembros, eliminar cambios que convergieron a un error, etc.

2. A continuación se detallan 3 de sus diferencias:

i. Naturaleza y propósito:

Git es un sistema de control de versiones distribuido que permite a los desarrolladores gestionar y rastrear cambios en el código fuente durante el desarrollo de software. Es una herramienta que se ejecuta localmente en la máquina del usuario.
GitHub es un servicio en línea que ofrece alojamiento para repositorios Git y proporciona funcionalidades adicionales para la colaboración, como el seguimiento de problemas, la revisión de código, y la integración continua, entre otros.

ii. Dependencia:

Git es independiente y no necesita de GitHub para funcionar. Se puede usar Git localmente o con otros servicios de alojamiento.
GitHub en cambio, se basa en Git. Es una plataforma que utiliza la tecnología de Git para proporcionar todas sus características relacionadas con el control de versiones.

iii. Interfaz:

Git opera principalmente a través de la línea de comandos, aunque existen interfaces gráficas disponibles para utilizar Git.
GitHub proporciona una interfaz web amigable para gestionar y visualizar repositorios, así como para colaborar con otros desarrolladores.

3. En Git, un hash es una suma de comprobación que le da un identificador único a los archivos en función de su contenido, de manera que no se pueda cambiar los contenidos del mismo, o su directorio, sin que Git pueda identificarlo. Este identificador es una cadena de cuarenta caracteres hexadecimales que es generado por un algoritmo matemático llamado SHA-1.
En un commit, el hash se utiliza para la identificación única de ese commit y rastrear su historia. Por ejemplo, si se quiere buscar un commit en específico donde se hizo X cambio, entonces se puede utilizar el hash de ese commit para encontrarlo.

4. `Git` utiliza una estructura de datos compleja para manejar el control de versiones. Estos son algunos de sus componentes principales:

**Blobs:** Representan el contenido de un archivo en un punto determinado en el tiempo. Cada blob es identificado por un hash SHA-1 de su contenido.

**Trees:** Representan directorios. Un árbol vincula nombres de archivo con blobs y puede referenciar otros árboles para representar subdirectorios.

**Commits:** Representan puntos en el tiempo o estados de un proyecto. Un commit apunta a un árbol, que representa el estado del proyecto en ese punto en el tiempo. Además, contiene metadatos como el autor, el mensaje del commit, y las referencias a commits padres (en caso de ser un merge, puede tener dos o más commits padres).

**Tags:** Son referencias a commits específicos, generalmente utilizados para marcar puntos importantes, como versiones de lanzamiento.

**Branches:** Son punteros móviles a un commit. Cuando se crea un nuevo commit, el puntero del branch se mueve automáticamente para apuntar a ese commit.

**HEAD:** Es un puntero especial que indica el commit actual donde está posicionado el repositorio. Por lo general, apunta al último commit de la rama en la que se está trabajando.

- git diff: Este comando muestra las diferencias en líneas agregadas, modificadas y eliminadas entre dos commits.

11. `git stash` (0.25 puntos)
12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

<code>Escriba sus respuestas acá:</code>

11. Este comando sirve para hacer un guardado temporal de los cambios realizados dentro del mismo directorio de trabajo. Esto se hace generalmente para cambiar de branch o de proyecto sin perder el progreso realizado, y sin necesidad de hacer un commit. Una vez se quiera volver a trabajar con los archivos guardados, se puede ejecutar el comando 'git stash list' para acceder a los archivos almacenados en stash. Por otro lado, también se puede utilizar para deshacer cambios almacenados en stash sin eliminar el commit original, o simplemente aplicar esos cambios en caso contrario.

12. **`git reset`**:
   
   **Uso**: El comando `git reset` se utiliza para restablecer el puntero HEAD a un commit específico. Puede ser usado de varias maneras dependiendo de las opciones utilizadas, y puede afectar tanto al índice (área de preparación) como al directorio de trabajo.

   - `git reset --soft <commit>`: Mueve el puntero HEAD al commit especificado, pero deja los cambios en el área de preparación. Es útil si se quiere rehacer un commit.
   - `git reset --mixed <commit>` (o simplemente `git reset <commit>`): Mueve el puntero HEAD al commit especificado y también deshace los cambios en el área de preparación, pero no en el directorio de trabajo.
   - `git reset --hard <commit>`: Mueve el puntero HEAD al commit especificado, y también deshace los cambios tanto en el área de preparación como en el directorio de trabajo.

   **Diferencia con `git revert`**:
   
   Aunque ambos comandos están relacionados con deshacer cambios, tienen diferencias fundamentales en su enfoque y efecto:
   
   - `git reset` retrocede el puntero HEAD (y opcionalmente el índice y el directorio de trabajo) a un commit anterior. Es una operación que reescribe la historia, y es potencialmente destructiva si se hace incorrectamente.
   
   - `git revert`, por otro lado, crea un nuevo commit que deshace los cambios realizados en otro commit. No modifica la historia; en cambio, agrega un nuevo commit al historial con las inversiones del commit que se está revirtiendo. Es una forma más segura de deshacer cambios si se está trabajando en una rama compartida, ya que no reescribe la historia del repositorio.

En resumen, `git reset` "retrocede" a un estado anterior, mientras que `git revert` deshace un cambio específico y lo registra en un nuevo commit.
